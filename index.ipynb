### Plot Categorical Variables

Now, pick 6 categorical variables and plot them against SalePrice with a bar graph for each variable. All 6 bar graphs should be on the same figure.

```python
cat_cols = list(ames.select_dtypes('object').columns)
cat_cols
```

```python
for col in cat_cols:
    print(col)
    print(ames[col].value_counts(dropna=False))
    print()
    print()
```

```python
chosen_cat = ['BldgType', 'HouseStyle', 'ExterQual', 'Foundation', 'KitchenQual', 'GarageType']
target = 'SalePrice'
```

```python
# Create bar plots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16,10), sharey=True)

for col, ax in zip(chosen_cat, axes.flatten()):
    (ames.groupby(col)        # group by column of interest
         .mean()[target]      # take the mean SalePrice of each group
         .sort_values()       # sort the groups in asc order of mean SalePrice
         .plot
         .bar(ax=ax))         # plot info on bar plot on separate ax
   
    ax.set_title(col)         # make the title of ea subplot the name of the col being graphed

fig.tight_layout()
```

## Create dummy variables

Create dummy variables for the six categorical features you chose remembering to drop the first. Drop the categorical columns that you used, concat the dummy columns to our continuous variables and asign it to a new variable `ames_preprocessed`

```python
ames_6cats = ames[chosen_cat]
ames_6cats.head()
```

```python
# Create dummy variables for your six categorical features
dummy_cats = pd.get_dummies(ames_6cats, drop_first=True)
dummy_cats.head()
```

```python
ames_preprocessed = ames.copy()
ames_preprocessed.drop(columns=chosen_cat, axis=1, inplace=True)
ames_preprocessed.head()
```

```python
ames_preprocessed = pd.concat([ames_preprocessed, dummy_cats], axis=1)
ames_preprocessed.head()
```
